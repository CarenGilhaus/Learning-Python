## 1. Satz des Pythagoras

Sie kennen den Satz des Pythagoras bereits aus der Schule: a² + b² = c². Lassen Sie uns nun eine Funktion ```pythagoras``` programmieren, die die beiden Katheten als Argumente ```a, b``` entgegennimmt und die Länge der Hypotenuse als Rückgabewert zurückgibt.

````python
# ERARBEITEN SIE IHRE LÖSUNG IN DIESER CODE-ZELLE
# - änderen Sie NICHT den untenstehenden Funktionsnamen
# - schreiben Sie Ihre Anweisungen dort, wo jetzt das Kommentar "YOUR CODE HERE" steht
# - sobald Sie ihre Anweisungen geschrieben haben, können Sie auch die Zeile "raise NotImplementedError()" löschen

def pythagoras(a,b):
    a = a * a
    b = b * b
    return((a+b)**(0.5))
````

````python
# CODE-ZELLE AUS ZUM TESTEN IHRER FUNKTION
# - die untenstehenden Ausdrücke vergleichen die Ausgaben Ihrer Funktion mit den von uns erwarteten Ergebnissen
# - führen Sie diese Zelle aus. Wenn drei mal True als Ausgabe unterhalb erscheint, ist Ihre Funktion korrekt!
# - wenn Sie stattdessen in der Ausgabe ein False erhalten, überprüfen Sie was den Fehler produziert haben könnte und ändern Sie ggf. Ihre Anweisungen

print(pythagoras(3,4) == 5.0)
print(pythagoras(5,12) == 13.0)
print(pythagoras(11,60) == 61.0)
````

````python
# ERARBEITEN SIE IHRE LÖSUNG IN DIESER CODE-ZELLE
# - änderen Sie NICHT den untenstehenden Funktionsnamen
# - schreiben Sie Ihre Anweisungen dort, wo jetzt das Kommentar "YOUR CODE HERE" steht
# - sobald Sie ihre Anweisungen geschrieben haben, können Sie auch die Zeile "raise NotImplementedError()" löschen

def salestax(vat_rate, net_price):
  return(net_price * (1+ vat_rate/100))


````
